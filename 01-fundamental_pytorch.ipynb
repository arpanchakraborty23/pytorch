# Tensor datatypes

```python
# float 32
float_32= torch.tensor([3.0,6.0,9.2],
                       dtype=None,
                        device= None,
                        requires_grad= False)
float_32
```

```python
float_32.dtype
```

```python
float_16= float_32.type(dtype=torch.float16)
float_16.dtype
```

```python
new_float=float_16 * float_32
new_float.dtype
```

```python
int_32_tensor= torch.tensor([3,4,6],dtype= torch.int32)
int_32_tensor
```

```python
int_32_tensor* float_32
```

```python
int_32_tensor * float_16
```

```python
int_32_tensor= torch.tensor([3,4,6],dtype= torch.long)
int_32_tensor
```

```python
int_32_tensor.dtype
```

```python
int_32_tensor * float_16
```
