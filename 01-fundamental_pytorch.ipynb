# tENSOR MANUPULATION

```python
# addition

rand_tensor= torch.rand(3,4)
rand_tensor
```

```python
rand_tensor + 10 
```

```python
# multiply
rand_tensor * 100
```

```python
# substraction
rand_tensor - 10.2
```

```python
# try torch in built
torch.mul(rand_tensor,5 )
```

```python
# matrix multiplecation
# dot product
matrix1= torch.rand(3,4)
matrix12= torch.rand(3,4)

torch.mul(matrix12,matrix1)
```

```python
%%time
v = 0
for i in range(len(rand_tensor)):
	v += rand_tensor[i] * rand_tensor[i]

print(v)
```

```python
%%time
tensor= torch.tensor([1,2,3])
torch.matmul(tensor, tensor)
```

```python
%%time
v = 0
for i in range(len(tensor)):
	v += tensor[i] * tensor[i]

print(v)
```
